# Cumulative reward plots

## Techniques
Nothing special, using pandas and json to parse a json file into a dataframe, then plotting it using matplotlib.

### Dependencies

```python
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt
```

### Reading the data

```python
with open("accumulated_reward_day.json") as f:
    items = json.load(f)["items"]
    df = pd.read_json(json.dumps(items), orient='columns', typ='frame', convert_dates=["day"])
#df = pd.read_json("C:/Users/Jaco/Documents/Python Scripts/Vizzing/accumulated_reward_day.json", orient='index', typ='frame')

first_row, df = df.iloc[0], df.iloc[1:]
display(df)
```

```python
plt.plot(df["accumulated_reward_amount"])
plt.show()
```

### Cumulative plot over time

```python
plt.figure(figsize=(20,10))
xdate = plt.plot_date(df["day"], df["accumulated_reward_amount"], xdate=True, ydate=False, linestyle='-', linewidth=3, marker='h', markeredgewidth=2, markerfacecolor='red', markeredgecolor='black', color='black', markersize=8)
plt.legend(['Accumulated Reward over time'], loc='upper left', prop={'size': 20})
#plt.plot_date(df["day"], np.zeros(len(df["day"])), markersize=0, linestyle='-', color='black')
plt.axvline(x='2018-08-14', color="black", linewidth=2)
plt.axvline(x=df['day'][1], color='black', linewidth=2)
plt.text(df['day'][1], 25000, str(df['day'][1])[:10] + ": Started keeping track", size=12, color='black')
plt.grid(b=True, which='major', axis='both')
plt.text('2018-07-01', -15000, "2018-08-14: Created a prediction Algorithm", size=12, color='black')
plt.savefig('AccumulatedRewardOverTime.png', dpi=1000)
plt.show()
```
